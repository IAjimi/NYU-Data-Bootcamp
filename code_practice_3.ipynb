### Question 3

Consider the 538 college majors data at url:
url1 = ’https://raw.githubusercontent.com/fivethirtyeight/data/master/’
url2 = ’college-majors/recent-grads.csv’
url = url1 + url2

```python
url1 = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/'
url2 = 'college-majors/recent-grads.csv'
url = url1 + url2
df538 = pd.read_csv(url)
```

a. Create a dataframe df538 from the csv file at url using read_csv(). What are its
dimensions?

```python
df538.shape
```

The dataframe has 173 rows and 21 columns.

b. What argument/parameter would you use to read only the first ten lines of the file?

```python
df538.head(n=10)
```

c. Extract the variables numbered [2, 4, 15, 16, 17]. What are the names of these
variables? What do they represent?

```python
df538.columns[[2,4,15,16,17]]
```

Those variables are 'Major' (the student's college major), 'Men' (the number of men in the major), 'Median' (the median yearly postgraduation job salary), 'P25th' (the 25th percentile of postgrad job salaries), 'P75th' (the 75th percentile of postgrad job salaries).

d. Set the index equal to Major.

```python
df538 = df538.set_index(['Major'])
```

e. Use the sort_values() method to sort the data by Total.

```python
df538.sort_values("Total")
```

f. What code would you use to extract the ten majors with the greatest number of people?

```python
df538["Total"].sort_values(ascending = False)[0:10]
```

alternatively, `df538["Total"].sort_values(ascending = False).head(10)`

g. Challenging. Construct horizontal bar charts of the top ten majors sorted, first, by median
salary and, second, by the salary of the 25th percentile. In each case plot just the variable
you sorted on.

```python
topmajors = df538.sort_values(by = "Total", ascending = False)[0:10]
topmajors["Median"].sort_values().plot(kind="barh")
plt.show()
```

```python
topmajors["P25th"].sort_values().plot(kind="barh")
plt.show()
```
